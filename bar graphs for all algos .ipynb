```python
import matplotlib.pyplot as plt
import numpy as np

accuracy_data = {
    'LR': 0.52,
    'KNN': 0.49,
    'SVM': 0.494,
    'RF': 0.498,
    'DT': 0.496,
    'NB': 0.495,
    'LSTM': 0.602,
    'MLP': 0.94,
    'XG Boost': 0.504
}

models = list(accuracy_data.keys())
accuracy_scores = list(accuracy_data.values())

# Find the index of the highest accuracy
max_index = accuracy_scores.index(max(accuracy_scores))

# Set colors, with 'royalblue' for the highest accuracy and a darker shade of blue for others
colors = ['royalblue' if i == max_index else '#73c2fb' for i in range(len(models))]

plt.figure(figsize=(10, 6))
plt.bar(models, accuracy_scores, color=colors)
plt.ylabel('Average Accuracy')
plt.xlabel('Algorithms')
plt.title('Overall Algorithm Comparison Across Datasets')
#plt.xticks(rotation=45, ha='right')

plt.show()

# Updated data for average
average_data = {
    'Logistic Regression': 0.52,
    'KNN': 0.49,
    'SVM': 0.494,
    'Random Forest': 0.498,
    'Decision Tree': 0.496,
    'Naïve Bayes': 0.495,
    'LSTM': 0.602,
    'MLP': 0.94,
    'XG Boost': 0.504
}

average_values = list(average_data.values())

# Print average values separated by commas
average_str = ', '.join(map(str, average_values))
print(f'average\n{average_str}')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Placeholder data for precision, f1 score, accuracy, and recall
precision_data = {
    'LR': 0.60,
    'ANN': 0.75,
    'KNN': 0.62,
    'SVM': 0.58,
    'RF': 0.59,
    'DT': 0.55,
    'NB': 0.48,
    'LSTM': 0.70,
    'MLP': 0.82
}

f1_score_data = {
    'LR': 0.45,
    'ANN': 0.68,
    'KNN': 0.50,
    'SVM': 0.48,
    'RF': 0.49,
    'DT': 0.47,
    'NB': 0.45,
    'LSTM': 0.60,
    'MLP': 0.78
}

accuracy_data = {
    'LR': 0.52,
    'ANN': 0.79,
    'KNN': 0.53,
    'SVM': 0.52,
    'RF': 0.52,
    'DT': 0.51,
    'NB': 0.50,
    'LSTM': 0.63,
    'MLP': 0.86
}

recall_data = {
    'LR': 0.55,
    'ANN': 0.72,
    'KNN': 0.58,
    'SVM': 0.54,
    'RF': 0.56,
    'DT': 0.52,
    'NB': 0.49,
    'LSTM': 0.68,
    'MLP': 0.80
}

models = list(precision_data.keys())
bar_width = 0.2
index = np.arange(len(models))

plt.bar(index, precision_data.values(), width=bar_width, color='blue', label='Precision')
plt.bar(index + bar_width, f1_score_data.values(), width=bar_width, color='green', label='F1 Score')
plt.bar(index + 2 * bar_width, accuracy_data.values(), width=bar_width, color='red', label='Accuracy')
plt.bar(index + 3 * bar_width, recall_data.values(), width=bar_width, color='purple', label='Recall')

plt.xlabel('Models')
plt.ylabel('Scores')
plt.title('Performance Metrics of Different Models')
plt.xticks(index + 1.5 * bar_width, models)
plt.legend()

plt.show()
```
