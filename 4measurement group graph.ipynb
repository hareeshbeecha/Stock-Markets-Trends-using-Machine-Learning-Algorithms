```python
df = pd.read_csv('modified_airtel.csv')
df.head()#importing the data set and printing it
```

```python
# Create predictor variables
df['Open-Close'] = df.Open - df.Close
df['High-Low'] = df.High - df.Low

# Store all predictor variables in a variable X
X = df[['Open-Close', 'High-Low']]
X.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you already have 'df' DataFrame with the 'Open-Close' and 'High-Low' columns
# If you haven't created these columns yet, use the code you provided in the previous message

# Create the plot
plt.figure(figsize=(10, 6))

# Plot 'Open-Close'
plt.plot(df['Date'], df['Open-Close'], label='Open-Close', color='blue')

# Plot 'High-Low'
plt.plot(df['Date'], df['High-Low'], label='High-Low', color='red')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Price Difference')
plt.title('Open-Close and High-Low Price Differences Over Time')

# Rotate x-axis date labels for better readability (optional)
plt.xticks(rotation=45)

# Show the legend
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
